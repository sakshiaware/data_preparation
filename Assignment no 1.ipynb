```python
y_pred=np.zeros(500)
y_actual=np.zeros(500)

for i in range (0,50):
    y_actual[i]=1
    
    for i in range(5,105):    
        y_pred[i]=1
```

```python
np.sum(y_actual)
```

```python
np.sum(y_pred)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
print("accuracy score:",accuracy_score(y_pred,y_actual))
```

```python
print("precision score:",precision_score(y_actual,y_pred))
```

```python
print(" recall score:",recall_score(y_actual,y_pred))
```

```python
print(" f1 score:",f1_score(y_actual,y_pred))
```

```python
from sklearn.metrics import confusion_matrix         #TN FP
confusion_matrix(y_actual,y_pred)                    #FN TP
```

```python
from sklearn.metrics import  classification_report
print(classification_report(y_actual,y_pred))
```
